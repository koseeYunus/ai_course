```python
# Bos verileri doldurmanin yollari: Ortalama, orta deger, en yakin komsularina bak, regression, en cok tekrar eden deger
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(strategy='mean') # strategy olarak ortalama secilir
```

```python
newData= imp.fit_transform(df) # df'deki verileri ortalama strategy ile verileri doldurma islemini yapar
```

```python
pd.DataFrame(newData) # yeni olusturulan array veriyi dataframe olarak gosterir
```

```python
from sklearn.impute import KNNImputer
```

```python
knn=KNNImputer()
```

```python
dolu=knn.fit_transform(df) # en yakin komsu ile verileri doldurur
```

```python
pd.DataFrame(dolu)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression() # Linear regrassion ile bos verileri doldurma
imp=IterativeImputer(estimator=lr)
lrdata=imp.fit_transform(df)
```

```python
pd.DataFrame(lrdata)
```
