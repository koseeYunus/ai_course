```python
# data imblanded(dengesiz)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x=df['text']
y=df['class']
```

```python
# smote=SMOTE(random_state=42)
# x_resample,yresample=smote.fit_resample(x,y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf=Pipeline([('CountVect',CountVectorizer()),('Classifier',RandomForestClassifier())])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.20, random_state = 42)
```

```python
clf.fit(x_train,y_train)
```

```python
tahmin=clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,tahmin)
```

```python
confusion_matrix(tahmin,y_test)
```

```python
from sklearn.svm import SVC
```

```python
svm=Pipeline([('CountVect',CountVectorizer(ngram_range=(1,2))),('Classifier',SVC())])
```

```python
svm.fit(x_train,y_train)
```

```python
tahmin2=svm.predict(x_test)
```

```python
accuracy_score(y_test,tahmin2)
```

```python
print(classification_report(y_test,tahmin2))
```

```python
confusion_matrix(tahmin2,y_test)
```

```python
sns.heatmap(confusion_matrix(tahmin2,y_test),annot=True)
```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def wc(data,color):
    plt.figure(figsize=(10,10))
    wc=WordCloud(background_color=color, stopwords=STOPWORDS)
    wc.generate(' '.join(data))
    plt.imshow(wc)
    plt.axis('off')
```

```python
spam=df[df['class']=='spam']
ham=df[df['class']=='ham']
```

```python
spam.head()
```

```python
wc(ham['text'],'white')
```

```python
wc(spam['text'],'red')
```

```python
def wc2(data,color):
    plt.figure(figsize=(10,10))
    mask=np.array(Image.open('cloud.png'))
    wc=WordCloud(background_color=color, stopwords=STOPWORDS,mask=mask)
    wc.generate(' '.join(data))
    plt.imshow(wc)
    plt.axis('off')
```

```python
wc2(ham['text'],'white')
```

```python
wc2(spam['text'],'red')
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline

model = AutoModelForSequenceClassification.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
tokenizer = AutoTokenizer.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
sa= pipeline("sentiment-analysis", tokenizer=tokenizer, model=model)

p = sa("bu telefon modelleri √ßok kaliteli , her par√ßasƒ± √ßok √∂zel bence")
print(p)
# [{'label': 'LABEL_1', 'score': 0.9871089}]
print(p[0]['label'] == 'LABEL_1')
# True

p = sa("Film √ßok k√∂t√º ve √ßok sahteydi")
print(p)
# [{'label': 'LABEL_0', 'score': 0.9975505}]
print(p[0]['label'] == 'LABEL_1')
# False
```

```python
p = sa("bugun hava guzel ama ogleden sonra yagmur yagma ihtimali varmis")
print(p)
print(p[0]['label'] == 'LABEL_1')
```
