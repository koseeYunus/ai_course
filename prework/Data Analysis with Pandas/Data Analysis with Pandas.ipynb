
#### Additional Dataframe Operations

16. Perform a `merge` using "client_list.csv" and "customer_id_list.csv". Assign the resulting dataframe as `clients`.
17. Perform a `merge` using `clients` and "purchase_log.csv" and limit the subset to only clients who made purchases. Assign the resulting dataframe as `detailed_sales`.
18. Use `groupby` to find the client who spent the most money on purchases. Determine how much he/she spent. HINT: save the intermediate dataframe from using `groupby` as `spenders` before applying slicing to determine the client who spent the most money on purchases.
19. (BONUS) Modify the answer to problem 3 slightly to determine exactly what items where purchased by the top spending client.

```python
# İki dosyayı birleştirme (merge)
client_list = pd.read_csv('client_list.csv')
customer_ids = pd.read_csv('customer_id_list.csv')
clients = pd.merge(client_list, customer_ids, 
                   left_on=['first_name', 'LAST_NAME'], 
                   right_on=['FIRST_NAME', 'LAST_NAME'],
                   how='inner')
clients.head()
```

```python
# Satın alma kayıtlarıyla birleştirme (sadece alışveriş yapanlar)
purchases = pd.read_csv('purchase_log.csv')
detailed_sales = pd.merge(clients, purchases, on='CUSTOMER_ID', how='inner')
detailed_sales.head()
```

```python
# En çok harcama yapan müşteriyi bulma
spenders = detailed_sales.groupby('CUSTOMER_ID')['PRICE'].sum().reset_index()
spenders.columns = ['CUSTOMER_ID', 'TOTAL_SPENT']
spenders = spenders.sort_values('TOTAL_SPENT', ascending=False)
top_spender = spenders.iloc[0]
print(f"En çok harcayan müşteri ID: {top_spender['CUSTOMER_ID']}")
print(f"Toplam harcama: ${top_spender['TOTAL_SPENT']:.2f}")
```

```python
# 19. BONUS: En çok harcayan müşterinin satın aldığı ürünler
top_customer_id = spenders.iloc[0]['CUSTOMER_ID']
top_customer_purchases = detailed_sales[detailed_sales['CUSTOMER_ID'] == top_customer_id]
print("\nEn çok harcayan müşterinin satın aldığı ürünler:")
print(top_customer_purchases[['ITEM_DESCRIPTION', 'PRICE', 'DATE']])
```
