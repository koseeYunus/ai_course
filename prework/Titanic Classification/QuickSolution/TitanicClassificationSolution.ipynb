```python
# Titanic - Keras Deep Learning Classification
# Hazırlayan: Yunus Köse (örnek notebook'lara göre sentez)
# Kaynak Notebook'lar:
# - TitanicClassificationSolution.ipynb
# - Normalization-PCA.ipynb
# - DataImputation.ipynb
# - DeepLearning.ipynb

# ======================================
# 1. Gerekli Kütüphaneler
# ======================================
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.decomposition import PCA

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam

# ======================================
# 2. Veri Yükleme
# ======================================
train = pd.read_csv("ttrain.csv")
test = pd.read_csv("ttest.csv")

print("Train shape:", train.shape)
print("Test shape:", test.shape)
train.head()

# ======================================
# 3. Veri Birleştirme (test verisini de birlikte dönüştürmek için)
# ======================================
y = train["Survived"]
train.drop("Survived", axis=1, inplace=True)
combined = pd.concat([train, test], axis=0, ignore_index=True)

# ======================================
# 4. Özellik Mühendisliği
# ======================================
combined["Title"] = combined["Name"].str.extract(" ([A-Za-z]+)\.")
combined["Title"] = combined["Title"].replace(
    ["Mlle", "Ms"], "Miss"
).replace(["Mme"], "Mrs")
rare_titles = combined["Title"].value_counts()[combined["Title"].value_counts() < 10].index
combined["Title"] = combined["Title"].replace(rare_titles, "Rare")

combined["FamilySize"] = combined["SibSp"] + combined["Parch"] + 1
combined["IsAlone"] = (combined["FamilySize"] == 1).astype(int)

combined["CabinLetter"] = combined["Cabin"].astype(str).str[0]
combined["Fare_log"] = np.log1p(combined["Fare"])

combined["TicketPrefix"] = combined["Ticket"].apply(lambda x: x.split(" ")[0] if not x.isdigit() else "NUM")

# Gereksiz sütunları at
combined.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1, inplace=True)

# ======================================
# 5. Sayısal ve Kategorik Sütunlar
# ======================================
num_features = combined.select_dtypes(include=["int64", "float64"]).columns
cat_features = combined.select_dtypes(include=["object"]).columns

# ======================================
# 6. Eksik Verilerin Doldurulması
# (DataImputation.ipynb'deki yöntemlere uygun şekilde)
# ======================================
num_pipeline = Pipeline([
    ("imputer", IterativeImputer(random_state=42)),
    ("scaler", StandardScaler())
])

cat_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer([
    ("num", num_pipeline, num_features),
    ("cat", cat_pipeline, cat_features)
])

# Dönüştürme işlemi
X_all = preprocessor.fit_transform(combined)

# ======================================
# 7. PCA (Normalization-PCA.ipynb'e uygun)
# ======================================
pca = PCA(n_components=0.95)  # varyansın %95'ini koru
X_pca = pca.fit_transform(X_all)

# Geri train-test ayırımı
X_train = X_pca[:len(y)]
X_test = X_pca[len(y):]

# ======================================
# 8. Train-Test Bölünmesi
# ======================================
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y, test_size=0.2, random_state=42)

# ======================================
# 9. Derin Öğrenme Modeli (DeepLearning.ipynb'e uygun)
# ======================================
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(X_tr.shape[1],)))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

history = model.fit(
    X_tr, y_tr,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=32,
    verbose=1
)

# ======================================
# 10. Performans Görselleştirme
# ======================================
plt.figure(figsize=(10,5))
plt.plot(history.history["accuracy"], label="Train Acc")
plt.plot(history.history["val_accuracy"], label="Val Acc")
plt.legend()
plt.title("Model Accuracy")
plt.show()

# ======================================
# 11. Test Verisi ile Tahmin
# ======================================
y_pred = (model.predict(X_test) > 0.5).astype(int).ravel()

submission = pd.DataFrame({
    "PassengerId": test["PassengerId"],
    "Survived": y_pred
})

submission.to_csv("submission_from_notebooks.csv", index=False)
print("✅ submission_from_notebooks.csv dosyası oluşturuldu.")
```
